# k-Means Clustering

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

## k-Means

El algoritmo de *k*-means busca por un pre-determinado numero de clusters multidimensionales en un conjunto de datos.
Cumpliendo con las siguientes caracteristicase:

- El "centro del cluster" es la mediana de todos los puntos pertenecientes al cluster.
- Cada punto asociado a su cluster esta mas cercano al suyo que a los otros.

Estos dos supuestos son la base d el modelo *k*-means model.

Generemos un conjunto de datos bidemensional que cotenga cuatro grupos de datos distintos.
OJO: esto es un algoritmo no supervisado por lo que dejamos las etiquetas fuera.

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Visualizando los resultados por colores (etiquetando).
Graficando tambien los centro de los clusters determinado por el estimador *k*-means:

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

El algoritmo *k*-means asign los puntos a los grupos de manera muy similar a como los podemos asignar con los ojos.
El número de posibles combinaciones de asignaciones de agrupamiento es exponencial en el número de puntos de datos, lo cual significa una busquedad exhaustiva.
El enfoque de *k*-means implica un enfoque iterativo intuitivo conocido como expectativa-maximizacióninvolves an intuitive iterative approach known as *expectation–maximization*.
